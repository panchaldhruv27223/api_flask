Q1. Explain GET and POST methods.

ANS:

GET :
    get and post both method is used for retriving data.
    using get method we retrive data from url,  in get our data stored in url so it is visible to everyone.
    this method is usefull when our data is not sensitive.
    get method can bookmarked,cached, it remain in the browser history.
    get method have length restriction.

POST :
    get and post both method is used for retriving data.
    using POST method we are not retrive data from url, we retrive data through form , our data is stored in http body not in 
    url so our data is not visible.
    this method is usefull when our data is sensitive like password , any algorithm key.
    post method can not bookmarked,cached, it is not remain in the browser history.
    post method have no-length restriction.
    it is a secure way to pass data.
