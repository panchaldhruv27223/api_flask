Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the 
url_for() function

```python
url_for() :- this function is used for url building.
    the first part of url_for() is a function and then we can pass any number of argument as we wish.
```

```python
from flask import Flask , url_for, redirect

app = Flask(__name__)

@app.route("/")
def welcome():
    return "<h1> welcome , just pass your age </h1>"

@app.route("/adult/<int:age>")
def adult(age):
    msg  = f"<h1> you are adult your age {age}  </h1>"

    return msg

@app.route("/not_adult/<int:age>")
def non_adult(age):
    msg  = f"<h1> you are not adult your age {age}  </h1>"

    return msg

@app.route("/check/<int:age>")
def chech_age(age):
    if age > 20:
        return redirect(url_for("adult",age=age))
    else:
        return redirect(url_for("non_adult",age=age))




if __name__ =="__main__":
    app.run(host="0.0.0.0" , debug=True)
```
